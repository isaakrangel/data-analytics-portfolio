**Loading the python library**

```python
# Necessary libraries 
import numpy as np
import pandas as pd
```

**Reading the data**

```python
# Loading the original data set and having a peek on the content

df = pd.read_csv('Exam_Score_Prediction.csv')
df.head(10)
```

```python
# Getting some info about the data, Dtype and shape

df.info()
```

**Formatting tables**

```python
# Replacing the undescores and white spaces to Empty spcaces, 

df.columns = df.columns.str.title().str.replace("_", "")
df.columns = df.columns.str.replace(" ", "")

# Removing the point from some categories in the course column

df['Course'] = df['Course'].str.replace('.', '', regex=False)

# Changing some categories names into a better format

df['StudyMethod'] = df['StudyMethod'].replace({
    'self-study': 'selfstudy',
    'online videos': 'online',
    'group study': 'groupstudy',
    'mixed': 'mixed'
})
df.head()
```

**Dropping StudentId column**

```python
df = df.drop('StudentId', axis=1) # Drop the ID column
```

# Converting some numerical features into categorical data

# ExamScore

```python
# Converting ExamScore values into grades, also creating a copy of the original dataset to keep the original data

df1 = df.copy() 
df1['Grade'] = df1['ExamScore'].apply(
    lambda x: 'A' if x >= 80 else
              'B' if x >= 70 else
              'C' if x >= 60 else
              'D' if x >= 50 else
              'F'
)
```

```python
# Dropping ExamScore column 
df1 = df1.drop(columns=['ExamScore'])
df1.head()
```

```python
df1['Grade'].value_counts()
```

# StudyHours

```python
# Converting StudyHours into levels 

df1['StudyLevel'] = df1['StudyHours'].apply(
    lambda x: 'low' if x <= 2 else ('medium' if x <= 5 else 'high')
)
```

```python
# Dropping StudyHours column 

df1 = df1.drop(columns=['StudyHours'])
df1.head()
```

# SleepHours

```python
df1['SleepLevel'] = df1['SleepHours'].apply(
    lambda x: 'bad' if x < 5 else ('regular' if x < 7 else 'good')
)
```

```python
# Dropping SleepHours column 

df1 = df1.drop(columns=['SleepHours'])
df1.head()
```

# ClassAttendance

```python
df1['AttendanceLevel'] = df1['ClassAttendance'].apply(
    lambda x: 'low' if x < 60 else ('medium' if x < 80 else 'high')
)
```

```python
# Dropping SleepHours column 

df1 = df1.drop(columns=['ClassAttendance'])
df1.head()
```

```python
# This save and creates a New file with the cleaned dataset

df1.to_csv('Exam_Score_Cleaned_Final.csv', index=False)
```
